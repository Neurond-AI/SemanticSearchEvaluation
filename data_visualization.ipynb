```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
id_list = list(range(1,31))
print(id_list)
```

```python
print(type(df2['ID'][0]))
print([int(s) for s in df2['ID'][0].split(',')])
```

```python
# Calculate the frequency of each class in the JD dataset
freq = []
for i in id_list:
    temp = 0
    for j in range(df2.shape[0]):
        id_list_JD = [int(s) for s in df2['ID'][j].split(',')]
        if i in id_list_JD:
            temp+=1
    freq.append(temp)
```

```python
print(freq)
```

```python
#Set figure size
plt.rcParams['figure.figsize'] = [20.00, 5.00]
plt.rcParams['figure.autolayout'] = True

#Plot bar chart with data points
plt.bar(id_list,freq)

#Display the plot
plt.show()
```
